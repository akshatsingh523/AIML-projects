```python
def alternate_values(column):
    alternating_values = [1, 0] * (len(column) // 2)
    if len(column) % 10 == 0:

        alternating_values.append(0)

    return alternating_values

# Apply the function to the specified columns
columns_to_alternate = ['BAD DEBT', 'PACKAGE DOWNGRADE', 'SR DONE','PACKAGE DONE','GA STATUS','PAYMENT SETTLED','Bill paid 1','Bill paid 2','Bill paid 3`']

for col in columns_to_alternate:
    df[col] = alternate_values(df[col])


# Print the resulting DataFrame
print(df)
```
